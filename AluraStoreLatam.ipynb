#1. Análisis de facturación



###En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingreso_total1 = tienda.Precio.sum()
ingreso_total2 = tienda2.Precio.sum()
ingreso_total3 = tienda3.Precio.sum()
ingreso_total4 = tienda4.Precio.sum()
print(f'Ingreso total tienda 1: {ingreso_total1}')
print(f'Ingreso total tienda 2: {ingreso_total2}')
print(f'Ingreso total tienda 3: {ingreso_total3}')
print(f'Ingreso total tienda 4: {ingreso_total4}')
ingreso_total = ingreso_total1 + ingreso_total2 + ingreso_total3 + ingreso_total4

```

```python

categorias = ['1','2', '3', '4']
valores = [ingreso_total1, ingreso_total2, ingreso_total3, ingreso_total4]

plt.figure(figsize=(7, 4))
plt.bar(categorias, valores, width=0.5)

plt.title("Ingresos Totales Por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Ventas (En millones)")
plt.grid(True)

plt.tight_layout()
plt.savefig('Ingresos_totales_por_tienda.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

def GraficoPastel(etiquetas,valores,numero_tienda):
  # Función personalizada para autopct que muestra el valor absoluto
  def autopct_valores_absolutos(val):
    a  = round(val/100.*sum(valores), 0)
    return f'{int(a)}' # Formatea el valor como un entero

  # Crear la figura y los ejes con un tamaño mayor
  fig, ax = plt.subplots(figsize=(8, 8))
  colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']



  # Crear el gráfico de pastel con colores diferentes
  ax.pie(valores, labels=etiquetas, autopct=autopct_valores_absolutos, colors=colores)
  ax.axis('equal')  # Asegura que el pastel sea un círculo perfecto

  plt.title(f'Ventas por Categoría de la tienda {numero_tienda}')

  plt.tight_layout()

  return fig
```

# 5. Envío promedio por tienda

### En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
promedio_envio_tienda1 = tienda['Costo de envío'].sum()/tienda['Costo de envío'].count()
promedio_alternativo = tienda['Costo de envío'].mean()
#print(f'Promedio de envío tienda 1: {promedio_envio_tienda1} y con el metodo alternativo {promedio_alternativo}')
print(f'Promedio de envío promedio para la tienda 1: {promedio_envio_tienda1}')

promedio_envio_tienda2 = tienda2['Costo de envío'].mean()
print(f'Promedio de envío promedio para la tienda 2: {promedio_envio_tienda2}')

promedio_envio_tienda3 = tienda3['Costo de envío'].mean()
print(f'Promedio de envío promedio para la tienda 3: {promedio_envio_tienda3}')

promedio_envio_tienda4 = tienda4['Costo de envío'].mean()
print(f'Promedio de envío promedio para la tienda 4: {promedio_envio_tienda4}')
```

#### Gráfico de Gasto promedio por envío Dependiendo de la tienda

```python
plt.style.use('_mpl-gallery')

# make data
x = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
y = [promedio_envio_tienda1, promedio_envio_tienda2, promedio_envio_tienda3, promedio_envio_tienda4]
# plot
fig, ax = plt.subplots(figsize=(5, 3)) # Aumentar el tamaño de la figura

ax.stem(x, y)


plt.title('Gasto promedio por envío por tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de envío promedio')
plt.grid(True)

plt.show()
```
